# 05. n-gram
与えられたシーケンス（文字列やリストなど）からn-gramを作る関数を作成せよ．この関数を用い，”I am an NLPer”という文から単語bi-gram，文字bi-gramを得よ．

```python
s = "I am an NLPer"

def n_gram(n, s):
    s_split = s.split(' ')
    word_gram = []
    char_gram = []
    for i in range(len(s_split)-(n-1)):
        word_gram.append(s_split[i:i+n])
    s = s.replace(' ','')
    for i in range(len(s)-(n-1)):
        char_gram.append(s[i:i+n])
    return word_gram, char_gram

n = 2
n_gram(n, s)    
```

# 08. 暗号文
与えられた文字列の各文字を，以下の仕様で変換する関数cipherを実装せよ．

英小文字ならば(219 - 文字コード)の文字に置換
その他の文字はそのまま出力
この関数を用い，英語のメッセージを暗号化・復号化せよ．

```python
def cipher(text):
    for c in text:
        if c.islower():
            print(chr(219-ord(c)),end='')
        else:
            print(c,end='')
            

cipher("aBcD")
```

# 41. 係り受け解析結果の読み込み（文節・係り受け）
40に加えて，文節を表すクラスChunkを実装せよ．このクラスは形態素（Morphオブジェクト）のリスト（morphs），係り先文節インデックス番号（dst），係り元文節インデックス番号のリスト（srcs）をメンバ変数に持つこととする．さらに，入力テキストの係り受け解析結果を読み込み，１文をChunkオブジェクトのリストとして表現し，冒頭の説明文の文節の文字列と係り先を表示せよ．本章の残りの問題では，ここで作ったプログラムを活用せよ．

```python
class Chunk():
    def __init__(self, line, temp):
        self.morphs:list = temp
        fileds = line.split(' ')
        self.dst:int = int(fileds[2].rstrip('D'))
        self.srcs:list = []
            
class Sentence():
    def __init__(self, morphs, chunks):
        self.morphs = morphs
        self.chunks = chunks
        for i, chunk in enumerate(self.chunks):
            if chunk.dst != -1:
                self.chunks[chunk.dst].srcs.append(i)
        

sentences:list = list()
morphs:list = list()
chunks:list = list()
temp:list = list() 
    
for line in lines:
    if line[0] == '*':
        if len(temp) > 0:
            chunks.append(Chunk(chunk, temp))
            chunk = line
            temp = []
        else:
            chunk = line
    elif line != 'EOS\n' and line != 'EOS':
        morph = Morph(line)
        morphs.append(morph)
        temp.append(morph)
    else:
        chunks.append(Chunk(chunk, temp))
        temp = []
        sentences.append(Sentence(morphs, chunks))
        morphs = []
        chunks = []
        
for sentenc in sentences[2].chunks:
    surface = [s.surface for s in sentenc.morphs]
    print(surface, sentenc.dst, sentenc.srcs)
```
